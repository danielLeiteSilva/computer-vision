```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # para exibir imagens dentro do COLAB
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
print('='*130 + '\n Passa baixa \n' + '='*130)
print('='*130 + '\n Original \n' + '='*130)
lena = cv.imread('Lena_ruido.jpg') #NÃ£o encontrei uma imagem boa com ruido             
lena = cv.cvtColor(lena, cv.COLOR_BGR2GRAY)
cv2_imshow(lena)

print('='*130 + 'Filtro Passa-baixa ' + '='*130)
kernel = np.ones((5,5),np.float32)/25         
lena_output = cv.filter2D(lena,-1,kernel) 
cv2_imshow(lena_output)
```

```python
print('='*130 + '\n Passa alta \n' + '='*130)
print('='*130 + '\n Imagem original em grey \n' + '='*130)
img_jupter = cv.imread('jupter.jpg')          
img_jupter = cv.cvtColor(img_jupter, cv.COLOR_BGR2GRAY) 
cv2_imshow(img_jupter)

print('='*130 + '\n filtro gradiente manual - Passa alta \n' + '='*130)
kernel = np.array([[ 1,  1,  1], [ 0,  0,  0],[-1, -1, -1]], np.float32)
img_jupter_output = cv.filter2D(img_jupter,-1,kernel) 
cv2_imshow(img_jupter_output)
```

```python
print("Derivativos")
img_input = cv.imread('jupter.jpg')    
img_input = cv.cvtColor(img_input, cv.COLOR_BGR2GRAY) 
derivativox = cv.Sobel(img_input,cv.CV_64F,1,0,ksize=5)
derivativoy = cv.Sobel(img_input,cv.CV_64F,0,1,ksize=5)

print('Derivativo X')
cv2_imshow(derivativox)      
print('Derivativo Y')
cv2_imshow(derivativoy)      
```
